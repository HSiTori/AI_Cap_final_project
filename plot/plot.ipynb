```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data1 = pd.read_csv('1_lr_high.txt', header=None, index_col=None)
data1.head()
```

```python
data2 = pd.read_csv('1_lr_low.txt', header=None, index_col=None)
data2.head()
```

```python
length = len(data2)
figure, ax1 = plt.subplots()
# np.arange(1, 10, 2)
ax1.plot(data1.iloc[1:length,0],data1.iloc[1:length,1],linewidth=1,zorder=1, label = "Force1")
ax1.plot(data1.iloc[1:length,0],data2.iloc[1:length+1,1],linewidth=1,zorder=1, label = "Force1")
ax1.legend(['learning rate = 0.04', 'learning rate = 0.00025'])
plt.ylabel('Average reward in 30 episodes')
plt.xlabel('Episodes')
plt.title('Learning rate comparison')
```

```python
from sklearn.preprocessing import normalize

data3 = pd.read_csv('5_lr_high.txt', header=None, index_col=None)
data3.head()
length = len(data2)
figure, ax1 = plt.subplots()
# np.arange(1, 10, 2)
data3.iloc[1:length,1] = data3.iloc[1:length,1] + 4;
data3.iloc[1:length,1] = data3.iloc[1:length,1] / np.linalg.norm(data3.iloc[1:length,1])
data1.iloc[1:length,1] = data1.iloc[1:length,1] / np.linalg.norm(data1.iloc[1:length,1])

ax1.plot(data1.iloc[1:length,0],data3.iloc[1:length,1],linewidth=1,zorder=1, label = "Force1")
ax1.plot(data1.iloc[1:length,0],data1.iloc[1:length,1],linewidth=1,zorder=1, label = "Force1")
ax1.legend(['5 lives', '1 live'])
plt.ylabel('Average reward in 30 episodes')
plt.xlabel('Episodes')
plt.title('Live number comparison')
```

```python
data4 = pd.read_csv('delay_lr_high.txt', header=None, index_col=None)
data5 = pd.read_csv('delay_lr_low.txt', header=None, index_col=None)
length = len(data2)
figure, ax1 = plt.subplots()

ax1.plot(data1.iloc[1:length,0],data4.iloc[1:length,1],linewidth=1,zorder=1, label = "Force1")
ax1.plot(data1.iloc[1:length,0],data5.iloc[1:length,1],linewidth=1,zorder=1, label = "Force1")
ax1.legend(['learning rate = 0.04', 'learning rate = 0.00025'])
plt.ylabel('Average reward in 30 episodes')
plt.xlabel('Episodes')
plt.title('Learning rate comparison with delay')
```
